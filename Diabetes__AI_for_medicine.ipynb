```python
import pandas as pd

# A public, de-identified dataset — no patient privacy concerns
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ["pregnancies", "glucose", "blood_pressure", "skin_thickness",
           "insulin", "bmi", "diabetes_pedigree", "age", "outcome"]
data = pd.read_csv(url, names=columns)

data.head()  # show the first 5 rows
```

```python
data.shape        # how many patients, how many variables
```

```python
data.info()       # data types, missing values
```

```python
data.describe()   # mean, min, max, etc. for each column
```

```python
import matplotlib.pyplot as plt

# Histogram of ages
data["age"].hist(bins=6)
plt.xlabel("Age")
plt.ylabel("Number of patients")
plt.title("Age distribution")
plt.show()
```

```python
# Compare glucose levels between diabetic and non-diabetic patients
data.boxplot(column="glucose", by="outcome")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Glucose level")
plt.title("Glucose by diabetes outcome")
plt.suptitle("")  # removes default subtitle
plt.show()
```

```python
# Scatter plot: BMI vs. Glucose, colored by outcome
plt.scatter(data["bmi"], data["glucose"], c=data["outcome"], cmap="coolwarm", alpha=0.6)
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.title("BMI vs Glucose (colored by diabetes outcome)")
plt.show()
```

```python
import seaborn as sns

sns.pairplot(data, hue="outcome", vars=["glucose", "bmi", "age"])
plt.show()
```
